```python
query = "Hello! I want to return my orders."

print("Original:", query)

cleaned = processor.clean_text(query)
print("Cleaned:", cleaned)

tokens = processor.tokenize(cleaned)
print("Tokens:", tokens)

filtered = processor.remove_stopwords(tokens)
print("Stopwords Removed:", filtered)

lemmas = processor.lemmatize(filtered)
print("Lemmatized:", lemmas)
```
